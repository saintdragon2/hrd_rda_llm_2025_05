# 인터넷에 있는 이미지 설명 요청하기

```python
from openai import OpenAI

client = OpenAI()

messages = [
    {
        "role": "system",
        "content": "너는 이미지에 대해 설명하는 AI야. "
    },
    {
        "role": "user",
        "content": [
            {"type": "text", "text": "이 이미지를 설명해줘. 어떤 작물인가?"},
            {
                "type": "image_url",
                "image_url": {
                    "url": "https://images.unsplash.com/photo-1499529112087-3cb3b73cec95?q=80&w=1548&auto=format&fit=crop&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D"
                }
            }
        ]
       
    }
]

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
    temperature=0.7
)

response.choices[0].message.content
```

# 내 컴퓨터에 있는 이미지 설명 요청하기

```python
# base64 인코딩

import base64

def encode_image(image_path): 
    with open(image_path, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    return encoded_string

image_path = "./data/damaged_apples.jpeg"
base64_image = encode_image(image_path)
print(base64_image)
```

```python

messages = [
    {
        "role": "system",
        "content": "너는 농업 관련 이미지에 대해 설명하는 AI야. 품종과 상태에 대해 자세히 설명하라."
    },
    {
        "role": "user",
        "content": [
            {"type": "text", "text": "이 이미지를 설명해줘. 어떤 품종인가?"},
            {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/jpeg;base64,{base64_image}"
                }
            }
        ]
       
    }
]

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
    temperature=0.7
)

response.choices[0].message.content
```
