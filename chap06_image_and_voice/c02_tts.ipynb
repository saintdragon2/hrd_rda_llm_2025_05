# TTS
- TTS: text-to-speech
- 문자를 음성으로 변환

```python
from openai import OpenAI

client = OpenAI()

response = client.audio.speech.create(
    model='tts-1-hd',
    voice='alloy', # 'ash', 'coral'
    input='안녕하세요. 반갑습니다. 농촌진흥청 농촌인적자원개발센터 교육에 오신 여러분 환영합니다.'
)

response.write_to_file('./data/hello.mp3')
```

```python
import IPython.display as ipd

ipd.Audio('./data/hello.mp3')
```

```python
response = client.audio.speech.create(
    model='tts-1-hd',
    voice='coral', # 'ash', 'coral'
    input='Hello. Nice to meet you. Welcome to the training of the Human Resource Development Center of the Rural Development Administration.'
)

response.write_to_file('./data/hello_coral.mp3')

ipd.Audio('./data/hello_coral.mp3')
```
