# TTS
- TTS: text-to-speech
- 문자를 음성으로 변환

```python
from openai import OpenAI

client = OpenAI()

response = client.audio.speech.create(
    model='tts-1-hd',
    voice='alloy', # 'ash', 'coral'
    input='안녕하세요. 반갑습니다. 농촌진흥청 농촌인적자원개발센터 교육에 오신 여러분 환영합니다.'
)

response.write_to_file('./data/hello.mp3')
```

```python
import IPython.display as ipd

ipd.Audio('./data/hello.mp3')
```

```python
response = client.audio.speech.create(
    model='tts-1-hd',
    voice='coral', # 'ash', 'coral'
    input='Hello. Nice to meet you. Welcome to the training of the Human Resource Development Center of the Rural Development Administration.'
)

response.write_to_file('./data/hello_coral.mp3')

ipd.Audio('./data/hello_coral.mp3')
```

```python
import base64

def encode_image(image_path): 
    with open(image_path, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    return encoded_string

image_path = "./data/damaged_apples.jpeg"
base64_image = encode_image(image_path)
print(base64_image)
```

```python

messages = [
    {
        "role": "system",
        "content": "너는 농업 관련 이미지에 대해 설명하는 AI야. 품종과 상태에 대해 자세히 설명하라."
    },
    {
        "role": "user",
        "content": [
            {"type": "text", "text": "이 이미지를 설명해줘. 어떤 품종인가?"},
            {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/jpeg;base64,{base64_image}"
                }
            }
        ]
       
    }
]

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
    temperature=0.7
)

answer = response.choices[0].message.content
print(answer)
```

```python
response = client.audio.speech.create(
    model='tts-1-hd',
    voice='coral', # 'ash', 'coral'
    input=answer
)

response.write_to_file('./data/damaged_apples.mp3')

ipd.Audio('./data/damaged_apples.mp3')
```
